```python
# Step 1: Import libraries


import pandas as pd
from sklearn.datasets import load_diabetes

from sklearn.model_selection import train_test_split
from sklearn.linear_model import *
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.metrics import *

df=sns.load_dataset('iris')
df

X=df.drop('species', axis=1)
X
y=df['species']


Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.4, random_state=10)
model = LogisticRegression()
model.fit(Xtrain, ytrain)
print(df.count())
print(Xtrain.count())
print(Xtest.count())
y_pred = (model.predict(Xtest))
r2 = model.score(Xtest, ytest)
print(f'R-squared: {r2:.2f}')

comparison_df = pd.DataFrame({'Actual': ytest, 'Predicted': y_pred})
print(comparison_df)
confusion_matrix = pd.crosstab(comparison_df['Actual'], comparison_df['Predicted'], rownames=['Actual'], colnames=['Predicted'])
print(confusion_matrix)
classification_report = classification_report(ytest, y_pred)
print(classification_report)
```
