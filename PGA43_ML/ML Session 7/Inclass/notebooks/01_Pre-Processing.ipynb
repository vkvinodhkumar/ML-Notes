```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import LabelEncoder
df=sns.load_dataset('iris')

data=pd.read_csv(r'D:\Git\Personal\PGA43_ML\ML Session 7\Inclass\data\online_gaming_behavior_dataset (1).csv')
le=LabelEncoder()
for col in data.select_dtypes(include=["object"]).columns:
    data[col]=le.fit_transform(data[col]).astype(str)

df['species_n']=le.fit_transform(df['species']).astype(str)
df
```

```python
print(data.head())
print("Information about the dataset:")
print(data.info())
print("Description of the dataset:")
print(data.describe(include='all'))
print("Checking for missing values:")
print(data.isnull().sum())
print("Checking for duplicate rows:")
print(data.duplicated().sum())
print("Checking for unique values in each column:")
for column in data.columns:
    print(f"{column}: {data[column].nunique()} unique values")
    print(f"{column}: {data[column].unique()} unique values")

print("Shape of the dataset:", data.shape)
```

```python
x=data.drop(['PlayerID','EngagementLevel'],axis=1)
print(x)
```

```python
y=data['EngagementLevel']
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
print(xtest.shape)
print(ytest.shape)
print(ytrain)
print(xtrain)
```

```python
train=pd.concat([xtrain,ytrain],axis=1)
train
```

```python
test=pd.concat([xtest,ytest],axis=1)
test
```

```python
train.to_csv('train.csv',index=False)
test.to_csv("test.csv",index=False)
```
